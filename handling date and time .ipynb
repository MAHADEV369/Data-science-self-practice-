```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#lets first extrac year
date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
#extracting month 
date['date_month_no'] = date['date'].dt.month
#naming them 
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python
#extracting days
```

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dateofweek'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week name

date['date_dayofweek_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#we can se if there is a weekend
date['date_is_weekend'] = np.where(date['date_dayofweek_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#extracting week if the year 
```

```python

date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
#extracting quarter
date['quarter'] = date['date'].dt.quarter
```
